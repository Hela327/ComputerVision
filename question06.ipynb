```python
#part b
import numpy as np

# Parameters
size = 5
sigma = 2

# Create coordinate grid (centered at 0)
ax = np.arange(-(size//2), size//2 + 1)
x, y = np.meshgrid(ax, ax)

# Gaussian function
G = (1/(2*np.pi*sigma**2)) * np.exp(-(x**2 + y**2)/(2*sigma**2))

# Derivatives
Gx = -(x / sigma**2) * G
Gy = -(y / sigma**2) * G

# Normalize (sum of absolute values = 1)
Gx = Gx / np.sum(np.abs(Gx))
Gy = Gy / np.sum(np.abs(Gy))

# Print kernels
print("Gx Kernel:\n", Gx)
print("\nGy Kernel:\n", Gy)
```

```python
#Part c
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameters
size = 51
sigma = 2

# Grid
ax = np.arange(-(size//2), size//2 + 1)
x, y = np.meshgrid(ax, ax)

# Gaussian
G = (1/(2*np.pi*sigma**2)) * np.exp(-(x**2 + y**2)/(2*sigma**2))

# Derivative (choose one: Gx or Gy)
Gx = -(x / sigma**2) * G   # x-direction

# 3D Plot
fig = plt.figure()
ax3d = fig.add_subplot(111, projection='3d')

ax3d.plot_surface(x, y, Gx, cmap='viridis')

ax3d.set_title("3D Surface of Derivative of Gaussian (Gx)")
plt.show()
```

```python
#part d
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('einstein.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Create DoG kernels (5x5, sigma=2)
size = 5
sigma = 2

ax = np.arange(-(size//2), size//2 + 1)
x, y = np.meshgrid(ax, ax)

G = (1/(2*np.pi*sigma**2)) * np.exp(-(x**2 + y**2)/(2*sigma**2))

Gx = -(x / sigma**2) * G
Gy = -(y / sigma**2) * G

# Normalize
Gx = Gx / np.sum(np.abs(Gx))
Gy = Gy / np.sum(np.abs(Gy))

# Apply kernels (convolution)
Ix = cv2.filter2D(gray, -1, Gx)   # horizontal gradient
Iy = cv2.filter2D(gray, -1, Gy)   # vertical gradient

# Show results
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(Ix, cmap='gray')
plt.title("Horizontal Gradient (Ix)")
plt.axis('off')



plt.show()
```

```python
plt.subplot(1,2,2)
plt.imshow(Iy, cmap='gray')
plt.title("Vertical Gradient (Iy)")
plt.axis('off')

plt.show()
```

```python
#part e
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('einstein.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Sobel gradients
sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5)
sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=5)

# Display
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(sobel_x, cmap='gray')
plt.title("Sobel X")
plt.axis('off')



plt.show()
```

```python
plt.subplot(1,2,2)
plt.imshow(sobel_y, cmap='gray')
plt.title("Sobel Y")
plt.axis('off')

plt.show()
```

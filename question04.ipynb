```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('woman.jpg')

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Otsu thresholding (INVERTED to get foreground)
T, binary_mask = cv2.threshold(
    gray, 0, 255, 
    cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU
)

# Print threshold value
print("Otsu Threshold Value:", T)

# Display results
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')



plt.show()
```

```python
plt.subplot(1,2,2)
plt.imshow(binary_mask, cmap='gray')
plt.title("Binary Mask (Foreground)")
plt.axis('off')

plt.show()
```

```python
#part b)
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('woman.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Otsu (foreground mask)
T, mask = cv2.threshold(
    gray, 0, 255, 
    cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU
)

# Histogram equalization (full image)
equalized_full = cv2.equalizeHist(gray)

# Apply ONLY to foreground
result = gray.copy()
result[mask == 255] = equalized_full[mask == 255]

# Show result
plt.imshow(result, cmap='gray')
plt.title("Foreground Equalized Image")
plt.axis('off')
plt.show()
```
